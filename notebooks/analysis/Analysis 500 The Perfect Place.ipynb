```python
#Returns "" for closed businesses, and the current current category for open businesses.
def given_cat_open(row, given_cat):
    if (row['is_open'] == 1 and given_cat in row['categories'].lower()):
        return True
    return False

def given_cat(row, given_cat):
    if (given_cat in row['categories'].lower()):
        return True
    return False

def given_cat_interactions(row, given_cat):
    if (row['is_open'] == 1 and given_cat in row['categories'].lower()):
        return row['interactionsWeighted']
    return 0
```

```python
businessPrepped['givenCat_and_open'] = businessPrepped.apply(lambda row : given_cat_open(row, given_category), axis=1)
businessPrepped['givenCat'] = businessPrepped.apply(lambda row :given_cat(row, given_category), axis=1)
businessPrepped['givenCat_interactions'] = businessPrepped.apply(lambda row: given_cat_interactions(row, given_category), axis = 1)
businessPrepped.head()
```
